```python
import pyarrow.parquet as pa
import numpy as np
```

```python
table = pa.read_table('data/train-00000-of-00001.parquet')
```

```python
df = table.to_pandas()
```

```python
df = df[df["text"].str.split().str.len() > 20]
```

```python
df["frag_1_len"] = np.random.randint(5,10, size = len(df))
df["frag_2_len"] = np.random.randint(5,10, size = len(df))
df["overlap_len"] = np.random.randint(0, 5, size = len(df))
```

```python
df["test_string_len"] = df["frag_1_len"] + df["frag_2_len"] - df["overlap_len"]
```

```python
def get_test_string(text, length):
    words = text.split()[:length]
    return " ".join(words)
```

```python
def get_frag_2(text, frag_len):
    words = text.split()[-frag_len:]
    return " ".join(words)
```

```python
df["test_string"] = df.apply(lambda x: get_test_string(x["text"], x["test_string_len"]), axis = 1)
```

```python
df["frag_1"] = df.apply(lambda x: get_test_string(x["text"], x["frag_1_len"]), axis = 1)
```

```python
df["frag_2"] = df.apply(lambda x: get_frag_2(x["test_string"], x["frag_2_len"]), axis = 1)
```

```python
save_df = df[["test_string", "frag_1", "frag_2"]]
```

```python
save_df.to_csv("training_pairs_df")
```
